# Data Exploration: SpreadViewer Integration Results

This notebook explores the data generated by the SpreadViewer integration script.

**Generated Data:**
- `de_m_08_25`: August 2025 German Monthly contract
- `de_m_09_25`: September 2025 German Monthly contract

Each contains:
- `trades.csv`: Individual trade data
- `orders.csv`: Order book data (bid/ask)
- `mid_prices.csv`: Mid-price time series
- `metadata.json`: Contract configuration

## 4. Comparative Analysis

```python
# Compare the two contracts
if trades_1 is not None and trades_2 is not None:
    print("ğŸ“Š CONTRACT COMPARISON:")
    print("=" * 50)
    
    comparison = pd.DataFrame({
        'DE_M_08_25 (Aug)': [
            trades_1.shape[0],
            orders_1.shape[0] if orders_1 is not None else 0,
            trades_1['price'].mean() if 'price' in trades_1.columns else np.nan,
            trades_1['price'].std() if 'price' in trades_1.columns else np.nan,
            trades_1['volume'].sum() if 'volume' in trades_1.columns else np.nan,
            trades_1.index.min(),
            trades_1.index.max()
        ],
        'DE_M_09_25 (Sep)': [
            trades_2.shape[0],
            orders_2.shape[0] if orders_2 is not None else 0,
            trades_2['price'].mean() if 'price' in trades_2.columns else np.nan,
            trades_2['price'].std() if 'price' in trades_2.columns else np.nan,
            trades_2['volume'].sum() if 'volume' in trades_2.columns else np.nan,
            trades_2.index.min(),
            trades_2.index.max()
        ]
    }, index=['Trades Count', 'Orders Count', 'Avg Price', 'Price Volatility', 'Total Volume', 'Start Date', 'End Date'])
    
    print(comparison)
```

## 5. Data Quality Checks

```python
# Data quality analysis
print("ğŸ” DATA QUALITY ANALYSIS:")
print("=" * 40)

datasets = [
    ('DE_M_08_25 Trades', trades_1),
    ('DE_M_08_25 Orders', orders_1),
    ('DE_M_09_25 Trades', trades_2),
    ('DE_M_09_25 Orders', orders_2)
]

for name, df in datasets:
    if df is not None:
        print(f"\nğŸ“Š {name}:")
        print(f"   ğŸ“ˆ Rows: {len(df):,}")
        print(f"   ğŸ“‹ Columns: {len(df.columns)}")
        print(f"   âŒ Missing values: {df.isnull().sum().sum():,}")
        print(f"   ğŸ”¢ Numeric columns: {len(df.select_dtypes(include=[np.number]).columns)}")
        print(f"   ğŸ“… Time span: {(df.index.max() - df.index.min()).days} days")
        
        # Check for duplicates
        duplicates = df.index.duplicated().sum()
        print(f"   ğŸ”„ Duplicate timestamps: {duplicates:,}")
        
        # Check for data gaps
        time_diffs = df.index.to_series().diff().dropna()
        median_gap = time_diffs.median()
        max_gap = time_diffs.max()
        print(f"   â±ï¸  Median time gap: {median_gap}")
        print(f"   â±ï¸  Max time gap: {max_gap}")
    else:
        print(f"\nâŒ {name}: Not available")
```
